```
import keras
from keras.models import Sequential
from keras.layers import Dense
from keras.optimizers import Adam
```

```
from random import randint
from sklearn.preprocessing import MinMaxScaler
import numpy as np
```

```
train_sample=[]
train_label=[]
for i in range(1000):
  young_age=randint(13,64)
  train_sample.append(young_age)
  train_label.append([0])

  old_age=randint(64,100)
  train_sample.append(old_age)
  train_label.append([1])
```

```
train_sample
```

```
train_sample=np.array(train_sample)
train_label=np.array(train_label)
from sklearn.preprocessing import MinMaxScaler
MN=MinMaxScaler(feature_range=(0,1))
mn_fit_trans=MN.fit_transform(train_sample.reshape(-1,1))
```

```
model=Sequential([Dense(16,input_dim=1,activation="relu"),Dense(32,activation="relu"),Dense(2,activation="softmax")])
```

```
model.compile(Adam(lr=0.01),loss='sparse_categorical_crossentropy',metrics=["accuracy"])
model.fit(train_sample,train_label,batch_size=10,epochs=10)
```

```
test_sample=[]
test_label=[]

for i in range(500):
  young_age=randint(13,64)
  test_sample.append(young_age)
  test_label.append(0)
  
  old_age=randint(64,100)
  test_sample.append(old_age)
  test_label.append(1)

  
```

```
test_sample=np.array(test_sample)
test_label=np.array(test_label)
```

```
test_pred=model.predict_classes(test_sample)
```

```
from sklearn.metrics import accuracy_score
accuracy_score(test_label,test_pred)
```
